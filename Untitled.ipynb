### To check for stationarity and apply differencing if necessary, we'll follow these steps:

#### 1-Check for Stationarity:
##### -We'll use the Augmented Dickey-Fuller (ADF) test to check if the time series is stationary. The null hypothesis of the ADF test is that the time series has a unit root, indicating it is non-stationary. A low p-value (typically  < 0.05) indicates the rejection of the null hypothesis, meaning the series is stationary

#### 2-Apply Differencing:
##### If the series is not stationary, we'll apply differencing to make it stationary. We'll continue differencing until the series becomes stationary.

```python
from statsmodels.tsa.stattools import adfuller

# perform ADF test on the 'Close' prices
result = adfuller(data['Close'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

### The p-value is much higher than 0.05, indicating that we fail to reject the null hypothesis. This means the time series is not stationary

```python
# Apply differencing to make the series stationary
data_diff = data['Close'].diff().dropna()

# Perform ADF test on the differenced series
result_diff = adfuller(data_diff)
adf_stat_diff = result_diff[0]
p_value_diff = result_diff[1]
Critical_Values_diff = result_diff[4]

adf_stat_diff, p_value_diff, Critical_Values_diff
```

### The p-value is now much lower than 0.05, indicating that we reject the null hypothesis. This means the differenced series is stationary

## Determine ARIMA parameters using ACF/PACF

### To determine the ARIMA parameters (p, d, q), we will use the Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots.

#### ● p (AutoRegressive part): Determined by the PACF plot, it indicates the lag beyond which the PACF cuts off.
#### ● d (Differencing part): We already determined this to be 1, as we made the series stationary by differencing once.
#### ● q (Moving Average part): Determined by the ACF plot, it indicates the lag beyond which the ACF cuts off.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# plot ACF and PACF
fig, axs = plt.subplots(2, figsize = (14, 10))

# ACF plot
plot_acf(data_diff, ax=axs[0], lags=50)
axs[0].set_title('Autocorrelation Function (ACF)')

# PACF plot
plot_pacf(data_diff, ax=axs[1], lags=50)
axs[1].set_title('Partial Autocorrelation Function (PACF)')

plt.show()
```

# Explanation

## Based on the ACF and PACF plots:

### ACF Plot: Shows the autocorrelation of the differenced series with various lags.

### The ACF cuts off after a few significant lags, which suggests the value of q.

#### PACF Plot: Shows the partial autocorrelation of the differenced series with various lags.

### The PACF cuts off after a few significant lags, which suggests the value of p.

#### From the plots, we can infer the following:

#### The ACF plot shows significant autocorrelation at lag 1, and then it diminishes. The PACF plot shows significant partial autocorrelation at lag 1, and then it diminishes. Suggested Parameters:

##### ● p (AR order): 1
##### ● d (Differencing order):1 (as determined earlier)
##### ● q (MA order): 1

### These parameters suggest that an ARIMA(1,1,1) model might be appropriate.

# Fitting ARIMA

```python
import statsmodels.api as sm

# fit the ARIMA(1,1,1) model
model = sm.tsa.ARIMA(data['Close'], order=(1,1,1))
model_fit = model.fit()

# Display the model summary
print(model_fit.summary())
```

# Model Fit Statistics

### ● Log Likelihood(-9038.956):
#### This is a measure of model fit, with higher values indicating a better fit. However, it is generally used in comparison with other models.
### ● AIC (18083.912), BIC(18101.958), HQIC(18090.401):
#### These are information criteria used for model selection.Lower values indicate a better model. They penalize model complexity (number of parameters) to prevent overfitting. 

# Diagnostic Tests

#### ●Ljung-Box (L1) (Q) (0.01), Prob(Q) (0.93):

#### This test checks for the presence of autocorrelation on the residuals. A hifh p-value (above 0.05) indicates that there is no significant autocorrelation, suggesting that the model has captures the dependencies in the data adequately.

#### ● Jarque-Bera (JB) (587346.36), Prob(JB) (0.00):

#### This test checks whether the residuals follow a normal distribution. A low p-value (below 0.05) indicates that the resu=iduals are not notmally distributed.

#### ● Heteroskedasticity (H) (26.66), Prob(H) (two-sided) (0.00):

#### This test checks for constant variance in the residuals. A low p-value (below 0.05) indicates that the residuals exhibit heteroskedasticity (changing variance).

#### ● Skew (-0.76), Kurtosis (71.22):

#### These values measure the asymmetry and peakedness of the residuals. A skew value close to zero indicates symmetric residuals, while a high kurtosis value indicates heavy tails.

# Summary

## The ARIMA (1,1,1) model seems to fit the data reasonably well, as indicated by the significant AR and MA coefficients. 

#### However, the diagnostic tests suggest issues: The residuals are not normally distributes (Jarque-Bera test). There is evidence of heteroskedasticity in the residuals. The high kurtosis indicate that the residuals have heavy tails, meaning there are more extreme values than expected in a normal distribution.

```python
# Plot residuals 
residuals = model_fit.resid
fig, axs = plt.subplots(3, figsize=(14, 12))

# Plot residuals 
axs[0].plot(residuals)
axs[0].set_title('Residuals')
axs[0].grid(True, which='both', linestyle='--', linewidth=0.5)

# ACF of residuals
plot_acf(residuals, ax = axs[1], lags=50)
axs[1].set_title('ACF of Residulas')

# PACF of residuals
plot_pacf(residuals, ax=axs[2], lags=50)
axs[2].set_title('PACF of Residuals')

plt.show()
```
